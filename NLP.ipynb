```python
#Tokenization with Keras
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(["I love Google AI", "AI is the future"])
print(tokenizer.word_index)
```

```python
 #Import the Tokenizer
from tensorflow.keras.preprocessing.text import Tokenizer

Tokenizer is a Keras utility used in Natural Language Processing (NLP) to convert text into numerical form so that machine learning models can process it.

2. Create a Tokenizer
tokenizer = Tokenizer(num_words=100)
Creates a tokenizer object.
num_words=100 means the tokenizer will keep track of the top 100 most frequent words.
Since our text contains far fewer than 100 words, all words will be included.
3. Fit the Tokenizer on Text
tokenizer.fit_on_texts([
    "I love Google AI",
    "AI is the future"
])

The tokenizer scans the sentences and:

Splits them into words.
Converts words to lowercase by default.
Counts how often each word appears.
Assigns each unique word an integer index.

Words found:

i
love
google
ai
is
the
future

Frequency count:

Word	Count
ai	2
i	1
love	1
google	1
is	1
the	1
future	1
4. Display Word Index
print(tokenizer.word_index)

Possible output:

{
 'ai': 1,
 'i': 2,
 'love': 3,
 'google': 4,
 'is': 5,
 'the': 6,
 'future': 7
}
Why is ai assigned 1?

Because it appears twice, making it the most frequent word.

The tokenizer generally assigns:

Lower numbers to more frequent words.
Higher numbers to less frequent words.
Converting Text to Numbers

After fitting, you can convert text into sequences:

print(tokenizer.texts_to_sequences(["Google AI is amazing"]))

Output:

[[4, 1, 5]]

Explanation:

Word	Index
google	4
ai	1
is	5
amazing	Not found

amazing is not in the vocabulary, so it is ignored.

Why Tokenization is Needed

Machine learning models cannot understand raw text:

"I love AI"

Tokenization converts it into numbers:

[2, 3, 1]

These numerical sequences can then be fed into neural networks such as:

Recurrent Neural Networks (RNNs)
LSTMs
GRUs
Transformers
Visual Flow
Text
 ↓
"I love Google AI"

Tokenizer
 ↓
Split into words
 ↓
Assign indices
 ↓
{
 'ai':1,
 'i':2,
 'love':3,
 'google':4
}

Convert text to sequence
 ↓
[2, 3, 4, 1]

So, fit_on_texts() learns the vocabulary, and word_index stores the mapping from words to integers.
```
