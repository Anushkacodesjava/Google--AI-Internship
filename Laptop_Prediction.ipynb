```python
import os

folder = "/content/drive/MyDrive/All"

print(os.listdir(folder)[:50])
```

```python
import os

image_files = [f for f in os.listdir(folder) if f.lower().endswith(('.jpg','.jpeg','.png'))]

print("Total images:", len(image_files))
print("Sample files:", image_files[:10])
```

```python
import shutil
import os

# Define the path for the organized dataset, distinct from the source folder
dataset_path = "/content/drive/MyDrive/LaptopDataset"

# --- ADDED: Clean the dataset_path before populating it ---
# This ensures no old, potentially misnamed directories interfere with the new structure.
if os.path.exists(dataset_path):
    shutil.rmtree(dataset_path)
    print(f"Removed existing dataset directory: {dataset_path}")
# --- END ADDED ---

# Create the main dataset directory if it doesn't exist (will be created freshly now)
os.makedirs(dataset_path, exist_ok=True)

# Loop through the image files and organize them into label-specific subfolders
for file in image_files:
    try:
        # Refined label extraction logic:
        # 1. Split by underscore to get the primary brand name.
        # 2. Use os.path.splitext to ensure any potential file extension in that part is removed.
        # 3. Clean and capitalize the label.
        label_candidate = file.split("_")[0]
        label_candidate = os.path.splitext(label_candidate)[0]
        label = label_candidate.strip().capitalize()

        # Create a folder for the label within the dataset_path
        label_folder = os.path.join(dataset_path, label)
        os.makedirs(label_folder, exist_ok=True)

        # Define source and destination paths
        src = os.path.join(folder, file)
        dst = os.path.join(label_folder, file)

        # Copy the image to its respective label folder
        shutil.copy(src, dst)
    except Exception as e:
        # Catch any other potential errors during copying
        print(f"Error processing file '{file}': {e}")

print("Dataset organized into folders вњ”")
```

```python
for cls in os.listdir(dataset_path):
    print(cls, ":", len(os.listdir(os.path.join(dataset_path, cls))))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(224,224),
    batch_size=16,
    class_mode='categorical',
    subset='training'
)

val_data = datagen.flow_from_directory(
    dataset_path,
    target_size=(224,224),
    batch_size=16,
    class_mode='categorical',
    subset='validation'
)

print(train_data.class_indices)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(224,224,3)),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Conv2D(128,(3,3),activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Dense(train_data.num_classes,activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

class_names = list(train_data.class_indices.keys())

img_path = image_files[0]  # you can replace with any image

img = image.load_img(os.path.join(folder, img_path), target_size=(224,224))
img = image.img_to_array(img)
img = np.expand_dims(img, axis=0) / 255.0

pred = model.predict(img)

print("Predicted Laptop Brand:", class_names[np.argmax(pred)])
```
